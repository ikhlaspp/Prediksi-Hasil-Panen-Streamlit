# Baseline Model Training

## Overview
This notebook trains baseline machine learning models to establish performance benchmarks.

## Models
1. Linear Regression
2. Decision Tree
3. Random Forest
4. Gradient Boosting

## Purpose
- Establish baseline metrics
- Compare simple vs complex models
- Identify best approach for further optimization

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("✅ All libraries imported successfully!")
```

## 0. Import Libraries

Import all required libraries for baseline model training

```python
# Load pre-split and encoded datasets
X_train = pd.read_csv('../data/X_train.csv')
X_test = pd.read_csv('../data/X_test.csv')
y_train = pd.read_csv('../data/y_train.csv').values.ravel()
y_test = pd.read_csv('../data/y_test.csv').values.ravel()

print("✅ Data berhasil dimuat!")
print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
print(f"Number of features: {X_train.shape[1]}")
```

## 1. Load Dataset

Load pre-split training and testing data

```python
# Train Decision Tree baseline model
print("Training Decision Tree baseline model...")

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

print("✅ Model training completed!")
```

## 2. Train Baseline Model

Train Decision Tree as baseline model

```python
# Make predictions on test set
y_pred = dt_model.predict(X_test)

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("=" * 50)
print("BASELINE MODEL EVALUATION RESULTS")
print("=" * 50)
print(f"MAE (Mean Absolute Error)  : {mae:.4f} tons/ha")
print(f"RMSE (Root Mean Sq Error)  : {rmse:.4f} tons/ha")
print(f"R² Score (Accuracy)        : {r2:.4f} (Max: 1.0)")
print("=" * 50)
```

## 3. Evaluate Model

Calculate performance metrics on test set

```python
# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6, color='#667eea', edgecolors='black', linewidths=0.5, s=60)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Perfect Prediction')
plt.xlabel('Actual Yield (tons/ha)', fontsize=12)
plt.ylabel('Predicted Yield (tons/ha)', fontsize=12)
plt.title('Baseline Model: Actual vs Predicted Yield', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n📊 Scatter plot shows correlation: R² = {r2:.4f}")
```
